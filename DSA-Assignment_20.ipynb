Given a binary tree, your task is to find subtree with maximum sum in tree.

Examples:

Input1 : Â  Â  Â  

       1

     / Â  \

   2Â  Â  Â  3

  / \Â  Â  / \

4 Â  5Â  6 Â  7

Output1 : 28

As all the tree elements are positive, the largest subtree sum is equal to sum of all tree elements.

Input2 :

 Â  Â  Â  1

     /Â  Â  \

  -2Â  Â  Â  3

  / \Â  Â  /Â  \

4 Â  5Â  -6 Â  2

Output2 : 7

Subtree with largest sum is :

 -2

 / \

4 Â  5

Also, entire tree sum is also 7.

</aside>

```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def maxSubtreeSum(root):
    if root is None:
        return 0
    
    leftSum = maxSubtreeSum(root.left)
    rightSum = maxSubtreeSum(root.right)
    
    currentSum = leftSum + rightSum + root.value
    
    return currentSum

def findMaxSubtreeSum(root):
    if root is None:
        return 0
    
    leftMaxSum = findMaxSubtreeSum(root.left)
    rightMaxSum = findMaxSubtreeSum(root.right)
    
    currentSum = maxSubtreeSum(root.left) + maxSubtreeSum(root.right) + root.value
    
    return max(leftMaxSum, rightMaxSum, currentSum)


# Example usage:

# Create the binary tree from the given example
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
root.right.left = TreeNode(6)
root.right.right = TreeNode(7)

# Find the maximum subtree sum
maxSum = findMaxSubtreeSum(root)
print(maxSum)  # Output: 28
```

Construct the BST (Binary Search Tree) from its given level order traversal.

Example:

Input: arr[] = {7, 4, 12, 3, 6, 8, 1, 5, 10}

Output: BST:

            7

         /Â  Â  \

       4 Â  Â  12

     /Â  \ Â  Â  /

    3 Â  6Â  8

   / Â   / Â    \

 1Â    5  Â    10

</aside>

```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def constructBST(arr, start, end):
    if start > end:
        return None

    node = TreeNode(arr[start])
    index = start + 1

    while index <= end and arr[index] < node.value:
        index += 1

    node.left = constructBST(arr, start + 1, index - 1)
    node.right = constructBST(arr, index, end)

    return node

def printBST(root, space=0):
    if root is None:
        return

    # Increase distance between levels
    space += 5

    # Process right child first
    printBST(root.right, space)

    # Print current node with indentation
    print()
    for i in range(5, space):
        print(" ", end="")
    print(root.value)

    # Process left child
    printBST(root.left, space)


# Example usage:

# Level order traversal array
arr = [7, 4, 12, 3, 6, 8, 1, 5, 10]

# Construct the BST
root = constructBST(arr, 0, len(arr) - 1)

# Print the BST
printBST(root)
```

Given an array of size n. The problem is to check whether the given array can represent the level order traversal of a Binary Search Tree or not.

Examples:

Input1 : arr[] = {7, 4, 12, 3, 6, 8, 1, 5, 10}

Output1 : Yes

For the given arr[], the Binary Search Tree is:

            7

         /Â  Â  \

       4 Â  Â  12

     /Â  \ Â  Â  /

    3 Â  6Â  8

   / Â   / Â    \

Input2 : arr[] = {11, 6, 13, 5, 12, 10}

Output2 : No

The given arr[] does not represent the level order traversal of a BST.

</aside>

```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def isLevelOrderBST(arr):
    if len(arr) == 0:
        return "Yes"

    stack = [TreeNode(arr[0])]
    i = 1

    while i < len(arr):
        temp = stack[-1]

        if arr[i] < temp.value:
            temp.left = TreeNode(arr[i])
            stack.append(temp.left)
        else:
            while len(stack) > 0 and stack[-1].value < arr[i]:
                temp = stack.pop()
                if temp is not None and temp.value > arr[i]:
                    return "No"

            temp.right = TreeNode(arr[i])
            stack.append(temp.right)

        i += 1

    return "Yes"

# Example usage:

# Level order traversal array
arr1 = [7, 4, 12, 3, 6, 8, 1, 5, 10]
arr2 = [11, 6, 13, 5, 12, 10]

# Check if arr1 represents a level order traversal of a BST
result1 = isLevelOrderBST(arr1)
print(result1)  # Output: Yes

# Check if arr2 represents a level order traversal of a BST
result2 = isLevelOrderBST(arr2)
print(result2)  # Output: No
```
